# Data Formatting

```python
# Import data processing packages
import numpy as np
import csv
import matplotlib.pyplot as plt
import math
```

```python
files = ['Data/Utah/Weather/logan_10yr_weather.csv',
        'Data/Utah/Weather/ogden_10yr_weather.csv',
        'Data/Utah/Weather/provo_10yr_weather.csv',
        'Data/Utah/Weather/saltlakecity_10yr_weather.csv']

# Initialising variables to define the range of useful data
start_year = 2009
season_start = 11
season_end = 5

# Initialising empty lists to store hourly weather data
station = []
date = []
precip = []
temp = []
wind_speed = []
wind_direction = []
```

```python
# Removes erroneous characters from integers
def remove_s_char(x):
    if(str(x[-1]) == "s"):
        return str(x[:-1])
    else:
        return x
    
# Checks date range to ensure only winter data is stored
def check_in_season(date):
    if(start_year <= int(date[:3])):
        return False
    elif(int(date[5:6]) > season_end and int(date[5:6]) < season_start):
        return False
    else:
        return True
```

```python
# Reading weather data from file
def import_weather(filepath):
    # Variables used for tracking consistent increments of data
    hour = -1
    day = ""
   
   
    
    with open(filepath) as station_weather:
        weather = csv.DictReader(station_weather)
        for row in weather:
            if(check_in_season(row['DATE'])):
            
            
                # Remove rogue station data 
                if(row['REPORT_TYPE'] == 'FM-15' or row['REPORT_TYPE'] == 'AUTO'):
                    
                    
                    # Check for new day
                    if(row['DATE'][8:10] != day):
                        if(hour != 24):
                            print(str(row['DATE'][:10]) + " --- " + str(hour))
                        day = row['DATE'][8:10]
                        hour = 0

                    station.append(filepath[18:-17])

                    date.append(row['DATE'][:4] + "/" + row['DATE'][5:7] + "/" + row['DATE'][8:10])

                    # Extracting Temperature Data
                    # If no record is avaiable, assume temp stayed the same as previous hour
                    if(row['HourlyDryBulbTemperature'] == "" and row['HourlyWetBulbTemperature'] == ""):
                        temp.append(temp[-1])
                    elif(row['HourlyDryBulbTemperature'] == ""):
                        temp.append(int(remove_s_char(row['HourlyWetBulbTemperature'])))
                    elif(row['HourlyWetBulbTemperature'] == ""):
                        temp.append(int(remove_s_char(row['HourlyDryBulbTemperature'])))
                    else:
                        temp.append(int(remove_s_char(row['HourlyDryBulbTemperature'])) + int(remove_s_char(row['HourlyWetBulbTemperature'])) / 2)



                    # Extracting Precipitation Data
                    # If no record is available, 0 precip is recorded
                    if(row['HourlyPrecipitation'] == "" or row['HourlyPrecipitation'] == "T"):
                        precip.append(float(0))
                    else:
                        precip.append(float(remove_s_char(row['HourlyPrecipitation'])))



                    # Extracting Wind Direction Data
                    if(row['HourlyWindDirection'] == "" or row['HourlyWindDirection'] == "VRB"):
                        wind_direction.append(wind_direction[-1])
                    else:
                        wind_direction.append(int(row['HourlyWindDirection']))


                    # Extracting Wind Speed Data
                    if(row['HourlyWindSpeed'] == ""):
                        wind_speed.append(wind_speed[-1])
                    else:
                        wind_speed.append(int(remove_s_char(row['HourlyWindSpeed'])))

                    hour += 1
                
import_weather(files[0]) 
print("COMPLETE")
```
